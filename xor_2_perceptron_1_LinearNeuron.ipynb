```python
import numpy as np
#import tensorflow as tf

#defines the Euclidean scalar product  
def mult(x,y):
	s=0
	for i in range(0, len(x)):
		s=s+x[i]*y[i]
	return s

#defines the Heaviside step function
def H(x):
        if (x>=0):
                s=1
        else:
                s=0
        return s
#define the Perceptron class
class Perceptron:
    def __init__(self, inp, weight, bias):
        self.inp=inp
        self.weight=weight
        self.bias=bias

    def output(self):
        return(H(mult(self.weight,self.inp)+self.bias))

    def print_weight(self):
        print(self.weight)

    def print_bias(self):
        print(self.bias)
        
    def print_out(self):
        print(output(self))

#define a linear neuron
        
class Neuron_L:
    def __init__(self, inp, weight, bias):
        self.inp=inp
        self.weight=weight
        self.bias=bias
       
    def output(self):
        return(mult(self.weight,self.inp)+self.bias)

    def print_weight(self):
        print(self.weight)

    def print_bias(self):
        print(self.bias)

#define a NN function
def net(x, w1, w2, b1, b2, w, b):
    n11=Perceptron(x, w1, b1)
    n12=Perceptron(x, w2, b2)
    y =[n11.output(), n12.output()]
    n2=Neuron_L(y, w, b)
    return(n2.output())


#weights
w1=[5, 3]
w2=[-1.2, -1]
#biasses
b1=-1.5
b2=1.5
#the second layer is only one Linear neuron
w=[1, 1]
b=-1
#Provide training set
xx=[[0, 0], [0, 1], [1, 0], [1, 1]]  #set of inputs
zz=[0, 1, 1, 0]   #target values

ans1 = net(xx[0], w1, w2, b1, b2, w, b)
print(str(ans1))
ans2 = net(xx[1], w1, w2, b1, b2, w, b)
print(str(ans2))
ans3 = net(xx[2], w1, w2, b1, b2, w, b)
print(str(ans3))
ans4 = net(xx[3], w1, w2, b1, b2, w, b)
print(str(ans4))
```
